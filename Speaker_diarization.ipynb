```python
import scipy

def is_speech(audio, fs, frame_size):
    def audioSlice(audio, fs, frame_size):
        framesamp = int(frame_size*fs)
        hopsamp = int(frame_size*fs)
        X = scipy.array([audio[i:i+framesamp] for i in range(0, len(audio)-framesamp, hopsamp)])
        return X
    out = [vad.is_speech(np.int16(fr).tobytes(), fs) for fr in audioSlice(audio, fs, frame_size)]
    out = np.int16(out)
    counts = np.bincount(out)
    i = np.argmax(counts)
    return bool(i)
    
```

```python
import numpy as np
from python_speech_features import mfcc
import scipy.io.wavfile as wav

def find_mean_std(X_train):
    X_train_vector = np.asarray(())
    for x in X_train[:20]:
        vector = x
        if X_train_vector.size == 0:
            X_train_vector = vector
        else:
            X_train_vector = np.vstack((X_train_vector, vector))
    mean = np.mean(X_train_vector, axis=0)
    std = np.std(X_train_vector, axis=0)

    return mean, std



def normalize(data, mean, std):
    out = []
    for d in data:
        out += [(d - mean)/std]
    return out
    
    
def get_data_as_features(data_directory, num_class, win_len, win_step, num_cep, nfft):
    features1 = []
    features = np.asarray(())
    for i in range(num_class):
        audio_clip = data_directory + '{}.wav'.format(i+1)
        (rate, sig) = wav.read(audio_clip)
        feature = mfcc(sig, rate, win_len, win_step, num_cep, nfft = nfft)    
        features1 += [feature]

    features2 = []

    for i in range(30):
        audio_clip = data_directory + 'off/{}.wav'.format(i+1)
        (rate, sig) = wav.read(audio_clip)
        feature = mfcc(sig, rate, win_len, win_step, num_cep, nfft = nfft)   
        features2 += [feature]


    X_train = features1[:20] + features2[:20]
    X_test = features1[20:] + features2[20:]


    Y_train = np.array([0]*20+[1]*20)
    Y_test = np.array([0]*10+[1]*10)


    mean, std = find_mean_std(X_train)

    X_train = normalize(X_train, mean, std)
    X_test = normalize(X_test, mean, std)

    return X_train, X_test, Y_train, Y_test
```

```python
import soundfile as sf
import torch
from datasets import load_dataset
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor, Wav2Vec2Config

# load pretrained model
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")


librispeech_samples_ds = load_dataset("patrickvonplaten/librispeech_asr_dummy", "clean", split="validation")

# load audio
audio_input, sample_rate = sf.read(librispeech_samples_ds[0]["file"])

# pad input values and return pt tensor
input_values = processor(audio_input, sampling_rate=sample_rate, return_tensors="pt").input_values

# INFERENCE

# retrieve logits & take argmax
logits = model(input_values).logits
predicted_ids = torch.argmax(logits, dim=-1)

# transcribe
transcription = processor.decode(predicted_ids[0])

# FINE-TUNE

target_transcription = "A MAN SAID TO THE UNIVERSE I EXIST"

# encode labels
with processor.as_target_processor():
  labels = processor(target_transcription, return_tensors="pt").input_ids

# compute loss by passing labels
loss = model(input_values, labels=labels).loss
loss.backward()
```
